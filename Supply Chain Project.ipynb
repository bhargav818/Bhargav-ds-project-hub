```python
from category_encoders import BinaryEncoder
```

```python
br = BinaryEncoder()
df_encoded = br.fit_transform(df1)

df_encoded.head(1)
```

```python
df_encoded.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

df_encoded[['sale_price', 'market_price', 'rating']] = scaler.fit_transform(df_encoded[['sale_price', 'market_price', 'rating']])
```

```python
print(df_encoded['sale_price'])
```

```python
x = df_encoded.values
```

```python
from sklearn.cluster import KMeans
```

```python
#Elbow curve method
```

```python
wcss = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters = k, init = 'k-means++', max_iter = 100, random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.show()
```

```python
#Optimal value of k is 2 or 4
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for i in range(2, 15):
    kmeans = KMeans(n_clusters = i, random_state = 32)
    kmeans.fit(x)
    silhouette_avg = silhouette_score(x, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)
```

```python
plt.plot(range(2, 15), silhouette_scores, marker = 'o')

plt.xlim(2,15)
plt.xticks(range(2,15))

plt.show()
```

```python
#silhouette score for 2 is higher hence optimal value of k is 2
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 0.95)
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
kmeans = KMeans(n_clusters = 2,max_iter = 150, init = 'k-means++',random_state = 42)
kmeans_labels = kmeans.fit_predict(x_pca)  #start making the clusters
```

```python
# Visualization
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=kmeans_labels, cmap='rainbow', s=50, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='yellow', marker='o', s=100, label='Centroids')
plt.title('KMeans Clustering after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(kmeans.labels_)
```

```python
df_decoded = br.inverse_transform(df_encoded)
df_decoded = pd.DataFrame(df_decoded)
df_decoded.head(1)
df_decoded[['sale_price','market_price','rating']] = scaler.inverse_transform(df_encoded[['sale_price','market_price','rating']])
df_decoded.head(1)
df_decoded[['sale_price','market_price']] = np.exp(df_decoded[['sale_price','market_price']])
df_decoded.head(1)
df_decoded['cluster_id'] = kmeans_labels
df_decoded.head(1)
```

```python
result = df_decoded.groupby('cluster_id')['sale_price'].mean()
result

result1 = df_decoded.groupby('cluster_id')['market_price'].mean()
result1

result2 = df_decoded.groupby('cluster_id')['rating'].mean()
result2

cluster_info = pd.concat([result, result1,result2],axis =1 )
cluster_info.columns = ['avg_sale_price','avg_market_price','avg_rating']
cluster_info
```
