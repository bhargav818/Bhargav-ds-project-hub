```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./csgo.csv")
df
```

```python
#Too many columns and hence the model tend to overfit hence we shall use LDA since it is a supervised model
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
        print(le.classes_)
```

```python
df.isnull().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
df["round_winner"].value_counts()
```

```python
x = df.drop("round_winner", axis = 1)
y = df["round_winner"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train, y_train)
```

```python
lda.transform(x_test)
```

```python
#96 columns/features: provides the importance of each feature for each class
lda.coef_
```

```python
# values are close to zero, hence we gonna increase the magnitude of the values|
```

```python
lda_coef = np.exp(np.abs(lda.coef_))

lda_coef
```

```python
lda_coef = lda_coef.flatten() #to convert it to 1d array
```

```python
df1=pd.DataFrame({'columns':x.columns,'lda_coef':lda_coef})
df1
```

```python
#lDA is performed to obtain the above scores for each and every column
```

```python
df1 = df1.sort_values(by = 'lda_coef', ascending = False)

df1
```

```python
top_20 = df1.head(20)
top_20
```

```python
top_20.index
```

```python
imp_cols = top_20.index
```

```python
x_train
```

```python
x_train = x_train[: , imp_cols]

x_train
```

```python
x_test
```

```python
x_test = x_test[: , imp_cols]

x_test
```
